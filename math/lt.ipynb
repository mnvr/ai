```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-3 * math.pi, 3 * math.pi, 100)
y = np.cos(x) * 50
#plt.plot(x, y); plt.grid();
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y)
ax.grid(True)

ticks = np.arange(-3 * math.pi, 3 * math.pi + math.pi/2, math.pi / 2)
for t in ticks:
    ax.axvline(t, color="blue", linestyle="--", alpha=0.2)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Circle x + y = 1 (parametric points)
t = np.linspace(0, 2*np.pi)
ax.plot(np.cos(t), np.sin(t))

# Hyperbola x - y = 1 (parametric points, two branches)
u = np.linspace(-2, 2)
ax.plot(np.cosh(u), np.sinh(u), c="red")
ax.plot(-np.cosh(u), np.sinh(u), c="red")

ax.axis("square")
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.grid(True)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

t = np.linspace(0, 2*np.pi)
ax.plot(np.cos(t), np.sin(t), label="Circle x + y = 1")
tc = np.linspace(0, np.pi/4)
ax.fill(np.r_[0, np.cos(tc), 0], np.r_[0, np.sin(tc), 0], 
        alpha=0.8, label="Angle")

u = np.linspace(-2, 2)
ax.plot(+np.cosh(u), np.sinh(u), c="red", label="Hyperbola x - y = 1")
ax.plot(-np.cosh(u), np.sinh(u), c="red") # mirror branch
uc = np.linspace(0, 1)
ax.fill(np.r_[0, np.cosh(uc), 0], np.r_[0, np.sinh(uc), 0], c="red",
       alpha=0.8, label="Hyperbolic angle (area)")

ax.axis("square")
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.grid(True, alpha=0.5)
ax.legend()
```
